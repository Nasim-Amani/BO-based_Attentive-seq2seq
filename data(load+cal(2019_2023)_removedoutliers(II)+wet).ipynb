```python
import pandas as pd
import numpy as np


# Load the data
df = pd.read_csv("load+cal(2019-2023)_removedoutliers(II).csv")
df['DateTime'] = pd.to_datetime(df['DateTime'])
df.info()
```

```python
wet1= pd.read_csv("POWER_Point_Hourly_20190329_20231205_034d87N_048d53E_LST (1).csv")
wet2= pd.read_csv("POWER_Point_Hourly_20190329_20231205_034d87N_048d53E_LST - Copy.csv")
```

WET1

```python
# Convert YEAR, MO, DY, HR columns into a single DateTime column
wet1["DateTime"] = pd.to_datetime(wet1[['YEAR', 'MO', 'DY', 'HR']].astype(str).agg('-'.join, axis=1), format="%Y-%m-%d-%H")


# Drop the 'YEAR', 'MO', 'DY', 'HR' columns
wet1 = wet1.drop(columns=['YEAR', 'MO', 'DY', 'HR'])

# Move 'DateTime' to the first column
wet1 = wet1[['DateTime'] + [col for col in wet1.columns if col != 'DateTime']]


wet1.head()
```

wet2

```python
# Convert YEAR, MO, DY, HR columns into a single DateTime column
wet2["DateTime"] = pd.to_datetime(wet2[['YEAR', 'MO', 'DY', 'HR']].astype(str).agg('-'.join, axis=1), format="%Y-%m-%d-%H")


# Drop the 'YEAR', 'MO', 'DY', 'HR' columns
wet2 = wet2.drop(columns=['YEAR', 'MO', 'DY', 'HR'])

# Move 'DateTime' to the first column
wet2 = wet2[['DateTime'] + [col for col in wet2.columns if col != 'DateTime']]


wet2.head()
```

```python
wet = wet2.merge(wet1, on='DateTime')
#df3=newdf2.drop(columns=['day_of_week', 'month'])
wet.head()
```

```python
alldata = df.merge(wet, on='DateTime')
#df3=newdf2.drop(columns=['day_of_week', 'month'])
alldata.head()
```

```python
alldata.info()
```

```python

# Check for missing values in each column
missing_values = alldata.isnull().sum()

# Print the results
missing_values
```
